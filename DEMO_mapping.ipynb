## `mapping.py`

mapping.py  has four  functions  where  each  of  the  functions  can  be  called  after  using  import  ela.

```python
import ela
from ela import *
import folium
import pandas as pd
import json
import numpy as np
import IPython.display
```

```python
with open('./ela/data/us-states.json') as f:
    states = json.load(f)
with open('./ela/data/us-counties.json') as f:
    counties = json.load(f)
```

### 1.geojson_to_df()
* This function needs one argument which is GeoJSON object (loaded from us-states.json or us-counties.json)
* This function reads a geojson file into a pandas dataframe.

```python
ela.geojson_to_df(states)
```

```python
ela.geojson_to_df(counties)
```

### 2. prediction_map()
* Two inputs are provided for the function, GeoJSON object containing state or county boundaries, a string either gen or stor.
* This function then returns folium map layer showing the input geographic features colored according to their predicted energy types

```python
m = folium.Map(location = ela.get_latlon_from_zip(98105),
               width='80%', height='80%', zoom_start=4)
f = ela.prediction_map(states, 'gen')
m.add_child(f)
folium.LayerControl().add_to(m)
m
```

```python
m = folium.Map(location = ela.get_latlon_from_zip(98105),
               width='80%', height='80%', zoom_start=4)
f = ela.prediction_map(states, 'stor')
m.add_child(f)
folium.LayerControl().add_to(m)
m
```

```python
m = folium.Map(location = ela.get_latlon_from_zip(98105),
               width='80%', height='80%', zoom_start=5)
f = ela.prediction_map(counties, 'gen')
m.add_child(f)
folium.LayerControl().add_to(m)
m
```

```python
m = folium.Map(location = ela.get_latlon_from_zip(98105),
               width='80%', height='80%', zoom_start=4)
f = ela.prediction_map(counties, 'stor')
m.add_child(f)
folium.LayerControl().add_to(m)
m
```

### 3. facility_map()
* Two inputs are added to this function, a string containing the two letter abbreviation of the state, other either gen or stor.
* This function then returns a folium map layer containing a marker for each facility.

```python
m = folium.Map(location = ela.get_latlon_from_zip(98105),
               width='80%', height='80%', zoom_start=6)
f = ela.facility_map('WA', 'gen')
m.add_child(f)
folium.LayerControl().add_to(m)
m
```

```python
m = folium.Map(location = ela.get_latlon_from_zip(98105),
               width='80%', height='80%', zoom_start=7)
f = ela.facility_map('WA', 'stor')
m.add_child(f)
folium.LayerControl().add_to(m)
m
```

## `mapping.py`

mapping.py  has two  functions  where  each  of  the  functions  can  be  called  after  using  import  ela.

```python
import ela
from ela import *
import folium
import pandas as pd
import json
import numpy as np
import IPython.display
```

```python
with open('./ela/data/us-states.json') as f:
    states = json.load(f)
with open('./ela/data/us-counties.json') as f:
    counties = json.load(f)
```

### 1.geojson_to_df()
* This function needs one argument which is GeoJSON object (loaded from us-states.json or us-counties.json)
* This function reads a geojson file into a pandas dataframe.

```python
ela.geojson_to_df(states)
```

```python
ela.geojson_to_df(counties)
```

### 2. state_map()
* This function needs a single input, a two letter abbreviation of a state.
* This function generates a blank folium map centered at the input state.

```python
ela.state_map('WA')
```

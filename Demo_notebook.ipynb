## `ela.py`

The file `ela.py` has 8 functions.Each of the functions can be called after using `import ela`

```python
import ela
import pandas as pd
```

### 1) get_latlon_from_zip  
* This function needs only 1 input which is the zipcode.
* This function enables the user to input a zipcode and find the respective latitude and longitude for it. Eg:

```python
ela.get_latlon_from_zip(98105)
```

### 2) get_state_from_zip
* This function needs only 1 input which is the zipcode.
* This function enables the user to input a zipcode and find the respective state for it. Eg:

```python
ela.get_state_from_zip(98105)
```

### 3) get_closest_facility

* This function needs 2 inputs, a tuple of length equal to 2 and a string 'gen' or 'stor'
* This function enables the user to input a tuple of latitude and longitude and find the closest generation or storage facilitystate for it. Eg:

```python
ela.get_closest_facility((47.66377, -122.30118),'gen')
```

Alternatively the `get_latlon_from__zip` function can be called and that will provide the tuple of lat-lon.

```python
ela.get_closest_facility(ela.get_latlon_from_zip(98105),'stor')
```

### 4) get_state_breakdown

* This function needs 2 inputs, a state in the 2 letter abbreviation and a pandas dataframe 'gen' or 'stor'
* This function enables the user to find the statewise energy generation and storage breakdown. Eg:

```python
gen_data = pd.read_csv('./ela/data/generation_data.csv')
stor_data= pd.read_csv('./ela/data/storage_data.csv')
```

```python
ela.get_state_breakdown('WA',gen_data)
```

```python
ela.get_state_breakdown('WA',stor_data)
```

### 5) get_energy_breakdown

* This function needs 1 input,a pandas dataframe 'gen' or 'stor'.
* This function enables the user to find the energy generation or storage breakdown in the U.S. Eg:

```python
ela.get_energy_breakdown(gen_data)
```

```python
ela.get_energy_breakdown(stor_data)
```

### 6) state_type

* This function needs 1 input,a pandas dataframe 'gen' or 'stor'.
* This function enables the user to find the state energy break down for a given facility data.


```python
ela.state_type(gen_data)
```

```python
ela.state_type(stor_data)
```

### 7) get_predicted_type

* This function needs 2 inputs, a tuple of length equal to 2 for the lat-lon and a string 'gen' or 'stor'.
* This function enables the user to predict the type of generation or storage for that location. Eg:

```python
ela.get_predicted_type((47.66377, -122.30118),'gen')
```

```python
ela.get_predicted_type((47.66377, -122.30118),'stor')
```

### 8) graph_state_breakdown

* This function needs 2 inputs, a state in the 2 letter abbreviation and a pandas dataframe 'gen' or 'stor'
* This function enables the user to find the statewise energy generation or storage breakdown in a pie chart. Eg:

```python
ela.graph_state_breakdown('WA',gen_data)
```

```python
ela.graph_state_breakdown('WA',stor_data)
```

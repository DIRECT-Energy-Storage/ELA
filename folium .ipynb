```python
def gen_statewise(state):
    gen_df = ela.gen_data
    gen_df = gen_df[gen_df.state==state]
    levels, labels = pd.factorize(gen_df.type)
    lats = gen_df.lat
    lons = gen_df.lon
    colors = levels
    step = cm.StepColormap(['b','r','y','g','m','c','k'],vmin=0,vmax=7)
    f = folium.map.FeatureGroup()

    for lat, lon, color in zip(lats, lons, colors):
        f.add_child(
         folium.features.CircleMarker(
            [lat, lon],
            radius=3,
            color=None,
            fill_color=step(color)
            )
    )
        
    m= folium.Map([lats.mean(),lons.mean()], zoom_start=5)
    m.add_child(f)
    return m
```

```python
gen_statewise('CA')
```

```python
def stor_statewise(state):
    stor_df = ela.stor_data
    stor_df = stor_df[stor_df.state==state]
    levels, labels = pd.factorize(stor_df.type)
    lats = stor_df.lat
    lons = stor_df.lon
    colors = levels
    step = cm.StepColormap(['b','r','y','g','m','c','k'],vmin=0,vmax=7)
    f = folium.map.FeatureGroup()

    for lat, lon, color in zip(lats, lons, colors):
        f.add_child(
         folium.features.CircleMarker(
            [lat, lon],
            radius=3,
            color=None,
            fill_color=step(color)
            )
    )
        
    m= folium.Map([lats.mean(),lons.mean()], zoom_start=5)
    m.add_child(f)
    return m
```

```python
stor_statewise('New York')
```

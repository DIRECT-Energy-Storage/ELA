```python
import folium
import branca
import ela
import pandas as pd
import branca.colormap as cm

```

```python
def gen_statewise(state):
    gen_df = ela.gen_data
    gen_df = gen_df[gen_df.state==state]
    m=folium.Map()
    for index,rows in gen_df.iterrows():
        f = folium.map.FeatureGroup()
        if rows[4]=='SOLAR':
            f.add_child(folium.features.CircleMarker([rows[2],rows[3]],radius=3,fill_color='#f6cc0a')) #YELLOW
            m.add_child(f)
        elif rows[4]=='HYDRO':
            f.add_child(folium.features.CircleMarker([rows[2],rows[3]],radius=3,fill_color='#2B90F5')) #BLUE
            m.add_child(f)
        elif rows[4]=='NUCLEAR':
            f.add_child(folium.features.CircleMarker([rows[2],rows[3]],radius=3,fill_color='#b9341b')) #RED
            m.add_child(f)
        elif rows[4]=='BIOMASS':
            f.add_child(folium.features.CircleMarker([rows[2],rows[3]],radius=3,fill_color='#55b64e')) #GREEN
            m.add_child(f)
        elif rows[4]=='OIL':
            f.add_child(folium.features.CircleMarker([rows[2],rows[3]],radius=3,fill_color='#ff6700')) #ORANGE
            m.add_child(f)
        elif rows[4]=='COAL':
            f.add_child(folium.features.CircleMarker([rows[2],rows[3]],radius=3,fill_color='#fe7b89')) #PINK
            m.add_child(f)
        elif rows[4]=='GEOTHERMAL':
            f.add_child(folium.features.CircleMarker([rows[2],rows[3]],radius=3,fill_color='#d400ee')) #PURPLE
            m.add_child(f)
        elif rows[4]=='WIND':
            f.add_child(folium.features.CircleMarker([rows[2],rows[3]],radius=3,fill_color='#d9ffd8')) #MINT
            m.add_child(f)
        else:
            pass
    return m
   
```

```python
gen_statewise('IN')
```

```python
def stor_statewise(state):
    stor_df = ela.stor_data
    stor_df = stor_df[stor_df.state==state]
    m=folium.Map()
    for index,rows in stor_df.iterrows():
        f = folium.map.FeatureGroup()
        if rows[4]=='Electro-mechanical':
            f.add_child(folium.features.CircleMarker([rows[1],rows[2]],radius=3,fill_color='#f6cc0a')) #YELLOW
            m.add_child(f)
        elif rows[4]=='Electro-chemical':
            f.add_child(folium.features.CircleMarker([rows[1],rows[2]],radius=3,fill_color='#2B90F5')) #BLUE
            m.add_child(f)
        elif rows[4]=='Pumped Hydro Storage':
            f.add_child(folium.features.CircleMarker([rows[1],rows[2]],radius=3,fill_color='#b9341b')) #RED
            m.add_child(f)
        elif rows[4]=='Thermal Storage':
            f.add_child(folium.features.CircleMarker([rows[1],rows[2]],radius=3,fill_color='#55b64e')) #GREEN
            m.add_child(f)
    return m
```

```python
stor_statewise('Texas')
```
